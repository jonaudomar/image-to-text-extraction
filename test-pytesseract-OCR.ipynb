```python
from PIL import Image
import pytesseract

###affichage
import time
import sys
import random
```

```python
#image = Image.open("texte.jpg")
#image = Image.open("discours.jpg")

image = Image.open("image.png")
```

```python
gray_image = image.convert("L")
gray_image
```

```python
text = pytesseract.image_to_string(
    gray_image,
    lang="fra",
    config="--psm 6"
)
```

```python
def typewriter_by_lines(text, base_delay=0.03):
    lines = text.splitlines()
    for line in lines:
        for char in line:
            sys.stdout.write(char)
            sys.stdout.flush()
            time.sleep(base_delay)
        sys.stdout.write("\n")
        sys.stdout.flush()
        time.sleep(base_delay * 6)
```

```python
typewriter_by_lines(text)
```
